```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import skimage as ski

__image__ = r'../images/eight.tif'
```

```python
Img = ski.io.imread(__image__, as_gray=True)
Img_noise = ski.util.random_noise(Img, mode = 'gaussian', mean=0, var=0.005)

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(Img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Img_noise, cmap='gray')
plt.title('Image with Noise')
plt.axis('off')

plt.show()
```
