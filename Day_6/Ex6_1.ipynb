```python
import cv2
import numpy as np
import PIL
import matplotlib.pyplot as plt
```

```python
image = np.double(PIL.Image.open(r'../images/airplane.png', 'r'))
```

```python
h1 = np.double(np.ones([1,10])/ 10)
h2 = np.double(np.ones([10,1])/ 10)
h3 = np.double(np.ones([10, 10])/ 100) 
```

```python
# Apply the filter horizontal
filtered_image_1 = cv2.filter2D(image, -1, h1, borderType=cv2.BORDER_REPLICATE)

# Apply the filter vertical
filtered_image_2 = cv2.filter2D(image, -1, h2, borderType=cv2.BORDER_REFLECT)

# Apply the filter box
filtered_image_3 = cv2.filter2D(image, -1, h3, borderType=cv2.BORDER_WRAP)
```

```python
_, axs = plt.subplots(2, 2, figsize=(20, 20))

axs[0, 0].imshow(image, cmap='gray')
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')

axs[0, 1].imshow(filtered_image_1, cmap='gray')
axs[0, 1].set_title('Horizontal Filter')
axs[0, 1].axis('off')

axs[1, 0].imshow(filtered_image_2, cmap='gray')
axs[1, 0].set_title('Vertical Filter')
axs[1, 0].axis('off')

axs[1, 1].imshow(filtered_image_3, cmap='gray')
axs[1, 1].set_title('Box Filter')
axs[1, 1].axis('off')


plt.show()
```
