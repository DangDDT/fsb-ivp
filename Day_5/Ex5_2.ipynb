```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
__image__ = r'../images/moon.jpg'
# read image
image = cv2.imread(__image__)

gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
eq_img = cv2.equalizeHist(gray_img)

plt.figure(1, figsize=(10, 5))
plt.subplot(1, 2 , 1)
plt.imshow(gray_img, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2 , 2)
plt.imshow(eq_img, cmap='gray')
plt.title('Equalized Image')
plt.show()

```

```python
count, bins = np.histogram(image.ravel(), bins=256, range=[0, 256])

# Multiply bins by 255
bins *= 255

# Calculate histogram after equalization
count_eq, bins_eq = np.histogram(eq_img.ravel(), bins=256, range=[0, 256])

# display histogram before and after equalization
# before
plt.figure(2, figsize=(15, 5))
plt.subplot(1, 2 , 1)
plt.hist(gray_img.ravel(), 256, [0, 256])
plt.title('Histogram of Original Image')
plt.xlabel('# of pixels')
plt.xlim([0, 256])
plt.ylabel('Gray level')

# after
plt.subplot(1, 2, 2)
plt.bar(bins_eq[:-1], count_eq, width=np.diff(bins_eq)[0])
plt.title('Histogram after equalization')
plt.xlabel('Grey level')
plt.ylabel('# of pixels')

plt.show()
```
