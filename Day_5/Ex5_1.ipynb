```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# display histogram of an image with number of pixels and grey scale
__image__ = r'../images/brain.jpg'
# read image
image = cv2.imread(__image__)
# convert image to grey scale
grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# calculate histogram
hist = cv2.calcHist([grey], [0], None, [256], [0, 256])
# display histogram
plt.figure(figsize=(18, 8))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.plot(hist)
plt.xlim([0, 256])
plt.xlabel('Grey Scale')
plt.ylabel('Number of Pixels')
plt.title('Histogram of an Image')
plt.show()
```
