```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
Img = np.array(Image.open(r'../images/nature_color.jpg'))
```

```python
FS = 10 # Font size for caption
```

```python
for NumOfBit in range(1, 9):  # From 1 to 8 bits
    # Quantize to a given number of bits
    NumOfLevel = 2 ** NumOfBit
    LevelGap = 256 / NumOfLevel
    quantized_Img = np.uint8(np.floor(Img / LevelGap) * LevelGap)

    # Plot image
    plt.subplot(2, 4, NumOfBit)
    plt.imshow(quantized_Img)
    plt.axis('off')  # Hide axis
    
    # Image caption
    name = f'{NumOfBit}-bit' if NumOfBit == 1 else f'{NumOfBit}-bits'
    plt.title(name, fontsize=FS)

plt.tight_layout()
plt.show()
```
