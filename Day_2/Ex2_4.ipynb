```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
Img = Image.open(r'../images/nature_color.jpg')

# brighter images based on red green blue components
def brighter(Img):
    r, g, b = Img.split()
    r = r.point(lambda i: i * 2)
    g = g.point(lambda i: i * 2.5)
    b = b.point(lambda i: i * 3)
    return Image.merge('RGB', (r, g, b))

# darker images based on red green blue components
def darker(Img):
    r, g, b = Img.split()
    r = r.point(lambda i: i * 0.5)
    g = g.point(lambda i: i * 0.5)
    b = b.point(lambda i: i * 0.5)
    return Image.merge('RGB', (r, g, b))


# show image

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(Img)

plt.subplot(2, 2, 2)
plt.title('Brighter Image')
plt.imshow(brighter(Img))

plt.subplot(2, 2, 3)
plt.title('Darker Image')
plt.imshow(darker(Img))

plt.show()


```
