```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
Img = np.array(Image.open(r'../images/sunset.jpg'))
```

```python
# Increase brightness by adding a constant value to each color channel
I1 = np.clip(Img + np.array([50, 100, 150]), 0, 255).astype(np.uint8)

# Decrease brightness by subtracting a constant value from each color channel
I2 = np.clip(Img - np.array([50, 75, 100]), 0, 255).astype(np.uint8)
```

```python
FS = 20 # Font size caption

# Display the original, brighter, darker images
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.imshow(Img)
plt.title('Original Image', fontsize=FS)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(I1)
plt.title('Brighter Image', fontsize=FS)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(I2)
plt.title('Darker Image', fontsize=FS)
plt.axis('off')

plt.tight_layout()
plt.show()
```
