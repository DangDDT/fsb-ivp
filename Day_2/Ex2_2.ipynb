```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the test image and convert to a double
Img = np.array(Image.open(r'../images/tiger.jpg')).astype(np.float64) / 255
```

```python
# Brightness adjustment
scale = 0.4  # Scale factor
scaled_Img = Img * scale
```

```python
FS = 20 # Set the font size caption

# Display the original and dark images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(Img, vmin=0, vmax=1)
plt.title('Original Image', fontsize=FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(scaled_Img, vmin=0, vmax=1)
plt.title('Scaled Image', fontsize=FS)
plt.axis('off')

plt.tight_layout()
plt.show()
```
