```python
import skimage as ski
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = ski.io.imread(r'../images/binary_objects.jpg')
```

```python
se1 = ski.morphology.disk(5)
BW1 = ski.morphology.dilation(image, se1)
```

```python
se2 = ski.morphology.disk(15)
BW2 = ski.morphology.dilation(image, se2)
```

```python
se3 = ski.morphology.square(7)
BW3 = ski.morphology.dilation(image, se3)
```

```python
se4 = ski.morphology.rectangle(5,10)
BW4 = ski.morphology.dilation(image, se4)
```

```python
def line(length, degree):
    se = np.zeros((length, length))
    if degree == 0:
        se[int(length/2), :] = 1
    elif degree == 90:
        se[:, int(length/2)] = 1
    else:
        for i in range(length):
            se[i, int(i*np.tan(np.deg2rad(degree)))] = 1
    return se

se5 = line(10, 90)
BW5 = ski.morphology.dilation(image, se5)
plt.imshow(se5, cmap='gray')
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i, (ax, img, title) in enumerate(zip(axs.ravel(), [image, BW1, BW2, BW3, BW4, BW5], ['Original', 'Disk 5', 'Disk 15', 'Square 7', 'Rectangle 5x10', 'Line 10'])):
    ax.imshow(img, cmap='gray')
    ax.set_title(title)
    ax.axis('off')

plt.show()
```
