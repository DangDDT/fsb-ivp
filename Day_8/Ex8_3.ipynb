```python
import skimage as ski
import numpy as np
import matplotlib.pyplot as plt
```

```python
def line(length, degree):
    se = np.zeros((length, length))
    if degree == 0:
        se[int(length/2), :] = 1
    elif degree == 90:
        se[:, int(length/2)] = 1
    else:
        for i in range(length):
            se[i, int(i*np.tan(np.deg2rad(degree)))] = 1
    return se
```

```python
image = ski.io.imread(r'../images/circles.png')
```

```python
se1 = ski.morphology.square(45)
se2 = ski.morphology.disk(48)
se3 = ski.morphology.rectangle(35,45)
se4 = ski.morphology.diamond(45)
se5 = line(45, 45)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original')
for i, (se, title) in enumerate(zip([se1, se2, se3, se4, se5], ['Square 45', 'Disk 48', 'Rectangle 35x45', 'Diamond 45', 'Line 45 deg. 45'])):
    BW = ski.morphology.erosion(image, se)
    plt.subplot(2, 3, i+2), plt.imshow(BW, cmap='gray'), plt.title(title)
plt.show()
```
