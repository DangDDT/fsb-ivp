```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
```

```python
#a Flipping an image up-dơn and left right
# Load the image
Img = Image.open(r'../images/atrium.jpg')

# Perform the flip operations
Img_ud = ImageOps.flip(Img)  # Flip up to down
Img_lr = ImageOps.mirror(Img)  # Flip left to right
```

```python
# Font size for titles
FS = 20

# Display the images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(Img)
plt.title('Original Image', fontsize=FS)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(Img_ud)
plt.title('Up-Down Image', fontsize=FS)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(Img_lr)
plt.title('Left-Right Image', fontsize=FS)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#b Rotate an image with an angle theta in degree
# Load the image
Img = Image.open(r'../images/eight.png')

# Rotate the image by 90 degrees
Img_rot = Img.rotate(90)
```

```python
# Display the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(Img)
plt.title('Original Image', fontsize=FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Img_rot)
plt.title('Rotated Image', fontsize=FS)  # Updated title for clarity
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#c Cropping Image
# Load the image
Img = Image.open(r'../images/mandrill.tif')

# Cropping the image based on provided dimensions
xmin, ymin = 100, 150
width, height = 500 - xmin, 450 - ymin
Img_cropped = Img.crop((xmin, ymin, xmin + width, ymin + height))
```

```python
# Display the original and cropped images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(Img)
plt.title('Original Image', fontsize=FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Img_cropped)
plt.title('Cropped Image', fontsize=FS)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#d Zoom in an image with different interpolation methods
# Load the image
Img = Image.open(r'../images/bird.jpg')

# Zoom in by a factor of 2 with different interpolation methods
Img_z1 = Img.resize((Img.width * 2, Img.height * 2), Image.BICUBIC)  # Bicubic
Img_z2 = Img.resize((Img.width * 2, Img.height * 2), Image.NEAREST)  # Nearest neighbor
Img_z3 = Img.resize((Img.width * 2, Img.height * 2), Image.BILINEAR)  # Bilinear
```

```python
# Display the original and zoomed images
FS = 20  # Font size for titles

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(Img)
plt.title('Original Image', fontsize=FS)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(Img_z1)
plt.title('Bicubic', fontsize=FS)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(Img_z2)
plt.title('Nearest', fontsize=FS)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(Img_z3)
plt.title('Bilinear', fontsize=FS)
plt.axis('off')

plt.tight_layout()
plt.show()
```
