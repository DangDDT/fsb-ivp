```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load and convert the image to grayscale and normalize it
Img = np.array(Image.open(r'../images/quadnight.jfif').convert('L')) / 255.0
```

```python
# Number of sample images to generate for each averaging experiment
nImages = [1, 2, 8, 16, 64, 128]
FS = 20  # Font size for titles

plt.figure(figsize=(12, 8))

for i, N in enumerate(nImages, 1):
    avg_img = np.zeros(Img.shape)
    
    # Add Gaussian noise and accumulate
    for _ in range(N):
        noisy_Img = Img + np.random.normal(0, np.sqrt(0.4), Img.shape)
        avg_img += noisy_Img
    
    # Average the accumulated images
    avg_img /= N
    
    # Display each averaged image
    plt.subplot(2, len(nImages) // 2, i)
    plt.imshow(avg_img, cmap='gray')
    plt.title(f'{N} image(s)', fontsize=FS)
    plt.axis('off')


plt.tight_layout()
plt.show()
```
