```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load the image and convert it to a NumPy array
Img = np.array(Image.open(r'../images/lion.jpg'))
FS = 20  # Font size for titles
```

```python
# a.Identity Transformation
# Plot the Transformation Function
x = np.arange(256)
plt.figure(figsize=(6, 6))
plt.plot(x, x)  # Identity transformation
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.grid(True)
plt.title('Identity Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.show()
```

```python
# Display original and identity-transformed image
Img1 = x[Img]  # This is essentially a no-op for identity transformation

# Display the Original and Transformed (Identical) Image
FS = 20  # Font size for titles
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(Img, cmap='gray')
plt.title('Original', fontsize=FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Img1, cmap='gray')
plt.title('Adjusted', fontsize=FS)
plt.axis('off')
plt.show()
```

```python
#b Negative Transformation
# Plot the transformation function
x = np.arange(256)
plt.figure(figsize=(6, 6))
plt.plot(x, 255 - x)  # Negative transformation
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.grid(True)
plt.title('Negative Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.show()
```

```python
# Perform Negative transformation
Img2 = 255 - Img

# Display the Original and Negative Transformed Image
FS = 20  # Font size for titles
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(Img, cmap='gray' if Img.ndim == 2 else None)
plt.title('Original', fontsize=FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Img2, cmap='gray' if Img2.ndim == 2 else None)
plt.title('Adjusted', fontsize=FS)
plt.axis('off')
plt.show()
```

```python
#c Logarithmic Transformation
# Define the logarithmic transformation function
x = np.arange(256)
c = 255 / np.log(256)
y = c * np.log(x + 1)

# Plot the Logarithmic Transformation Function
plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.grid(True)
plt.title('Logarithmic Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.show()
```

```python
# Display the Original and Logarithmically Transformed Image
# Ensure Img is a numpy array before applying transformation
Img3 = np.uint8(y[Img])
FS = 20  # Font size for titles
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(Img, cmap='gray' if Img.ndim == 2 else None)
plt.title('Original', fontsize=FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Img3, cmap='gray' if Img3.ndim == 2 else None)
plt.title('Adjusted', fontsize=FS)
plt.axis('off')
plt.show()
```

```python
#d. Linear Piecewise Transformation
# Define the piecewise linear transformation function
y = np.zeros((256,), dtype=np.uint8)
y[:100] = np.linspace(0, 99, 100, dtype=np.uint8)
y[100:200] = 255
y[200:] = 150
```

```python
# Plot the Piecewise Linear Transformation Function
plt.figure(figsize=(6, 6))
plt.plot(np.arange(256), y)
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.grid(True)
plt.title('Piecewise Linear Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.axis('tight')
plt.axis('square')
plt.show()
```

```python
# Apply the piecewise linear transformation to the image
# For colored images, apply this transformation per channel
Img4 = y[Img]

# Display the Original and Transformed Image
FS = 20  # Font size for titles
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(Img, cmap='gray')
plt.title('Original', fontsize=FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Img4, cmap='gray')
plt.title('Adjusted', fontsize=FS)
plt.axis('off')
plt.show()
```

```python
#e. Exponential Transformation
# Define the exponential transformation function
c_exp = 255 / np.log(256)
x = np.arange(256)
y_exp = np.exp(x / c_exp) - 1

# Plot the Exponential Transformation Function
plt.figure(figsize=(6, 6))
plt.plot(x, y_exp)
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.grid(True)
plt.title('Exponential Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.show()
```

```python
# Define the nth root transformation function
n = 2  # Square root transformation
c_root = 255 / (255 ** n)
y_root = np.power(x / c_root, 1/n)

# Plot the Nth Root Transformation Function
plt.figure(figsize=(6, 6))
plt.plot(x, y_root)
plt.xlim([0, 255])
plt.ylim([0, 255])
plt.grid(True)
plt.title(f'{n}th Root Transformation Function')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.show()
```

```python
# Apply the Exponential Transformation
Img_exp = np.uint8(np.clip(y_exp[Img], 0, 255))

# Apply the Nth Root Transformation
Img_root = np.uint8(np.clip(y_root[Img], 0, 255))

# Display and save images as needed
FS = 20  # Font size for titles
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(Img_exp, cmap='gray')
plt.title('Exponential', fontsize=FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(Img_root, cmap='gray')
plt.title('Nth Root', fontsize=FS)
plt.axis('off')
plt.show()
```
