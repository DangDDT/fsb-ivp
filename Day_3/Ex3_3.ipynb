```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load test images
noText = Image.open(r'../images/gradient2.jpg')
withText = Image.open(r'../images/gradient_with_text2.jpg')
```

```python
# convert to array
notext_array = np.array(noText)
withtext_array = np.array(withText)
```

```python
# Divide the images
D = withtext_array / notext_array
threshold = 1
D = (D > threshold).astype(int) * 255
```

```python
# Plotting
FS = 20  # Font size for titles
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(withtext_array)
plt.title('Have Text', fontsize=FS)

plt.subplot(1, 3, 2)
plt.imshow(notext_array)
plt.title('No Text', fontsize=FS)

plt.subplot(1, 3, 3)
plt.imshow(D, cmap='gray')
plt.title('Detected Text', fontsize=FS)

plt.show()
```
