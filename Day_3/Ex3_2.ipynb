```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load test images
I1 = np.array(Image.open(r'../images/earth1.jpg')) / 255.0
I2 = np.array(Image.open(r'../images/earth2.jpg')) / 255.0

# Perform element-wise multiplication
I = np.multiply(I1, I2)
```

```python
# Plotting
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(I1)
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(I2)
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(I)
plt.title('Multiplication Image')
plt.axis('off')


plt.show()
```
