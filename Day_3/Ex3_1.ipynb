```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load images
Ia = Image.open(r'../images/prarie.jpg')
Ib = Image.open(r'../images/giraffe.jpg')
```

```python
# Resize Ib to match Ia's dimensions
Ib_resized = Ib.resize(Ia.size)
```

```python
# Convert images to numpy arrays for addition
Ia_array = np.array(Ia)
Ib_resized_array = np.array(Ib_resized)
# Perform addition
Ic_array = Ia_array + Ib_resized_array

# Convert back to an image
Ic = Image.fromarray(np.uint8(Ic_array))
```

```python
# Plotting
FS = 20  # Font size for titles
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(Ia)
plt.title('Image 1', fontsize=FS)

plt.subplot(1, 3, 2)
plt.imshow(Ib_resized)
plt.title('Image 2', fontsize=FS)

plt.subplot(1, 3, 3)
plt.imshow(Ic)
plt.title('Addition Image', fontsize=FS)

plt.tight_layout()
plt.show()
```
